```python
import pandas as pd
import matplotlib.pyplot as plt

path = 'bestsellers_with_categories.csv'

try:
    df = pd.read_csv(path)
    print("File successfully loaded!")
    
    df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
    
    plt.figure(figsize=(8, 6))
    df['price'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')
    plt.title('Price Distribution')
    plt.xlabel('Price (in dollars)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
    
    max_price = df['price'].max()
    min_price = df['price'].min()
    mean_price = df['price'].mean()
    median_price = df['price'].median()
    
    print(f"Maximum Price: ${max_price}")
    print(f"Minimum Price: ${min_price}")
    print(f"Average (Mean) Price: ${mean_price:.2f}")
    print(f"Median Price: ${median_price}")
    
except FileNotFoundError as e:
    print(f"Error: {e}")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

path = 'bestsellers_with_categories.csv'

try:
    df = pd.read_csv(path)
    print("File successfully loaded!")
    
    df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
    print("Column names have been updated to snake_case.\n")
    
    highest_rating = df['user_rating'].max()
    print(f"Highest rating in the dataset: {highest_rating}")
    
    count_highest_rating = df[df['user_rating'] == highest_rating].shape[0]
    print(f"Number of books with the highest rating: {count_highest_rating}")
    
    book_most_reviews = df.loc[df['reviews'].idxmax()]
    print(f"Book with the most reviews: {book_most_reviews['name']} by {book_most_reviews['author']}, with {book_most_reviews['reviews']} reviews.")
    
    books_2015 = df[df['year'] == 2015]
    most_expensive_2015 = books_2015.loc[books_2015['price'].idxmax()]
    print(f"The most expensive book in 2015 is '{most_expensive_2015['name']}' by {most_expensive_2015['author']}, costing ${most_expensive_2015['price']}.")
    
    fiction_2010 = df[(df['genre'] == 'Fiction') & (df['year'] == 2010)]
    fiction_2010_count = fiction_2010.shape[0]
    print(f"Number of Fiction books in the Top 50 in 2010: {fiction_2010_count}")
    
    books_4_9_rating = df[(df['user_rating'] == 4.9) & (df['year'].isin([2010, 2011]))]
    count_books_4_9 = books_4_9_rating.shape[0]
    print(f"Number of books with a rating of 4.9 in 2010 and 2011: {count_books_4_9}")
    
    books_2015_under_8 = books_2015[books_2015['price'] < 8]
    sorted_books_2015_under_8 = books_2015_under_8.sort_values(by='price')
    last_book_sorted = sorted_books_2015_under_8.iloc[-1]
    print(f"The last book in the sorted list is '{last_book_sorted['name']}' by {last_book_sorted['author']}, costing ${last_book_sorted['price']}.")
    
    plt.figure(figsize=(8, 6))
    df['price'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')
    plt.title('Price Distribution')
    plt.xlabel('Price (in dollars)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
    
    max_price = df['price'].max()
    min_price = df['price'].min()
    mean_price = df['price'].mean()
    median_price = df['price'].median()
    
    print(f"Maximum Price: ${max_price}")
    print(f"Minimum Price: ${min_price}")
    print(f"Average (Mean) Price: ${mean_price:.2f}")
    print(f"Median Price: ${median_price}")

except FileNotFoundError as e:
    print(f"Error: {e}")
```

```python
import pandas as pd

path = 'bestsellers_with_categories.csv'

df = pd.read_csv(path)

df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

price_stats = df.groupby('genre').agg(max_price=('price', 'max'), min_price=('price', 'min')).reset_index()

fiction_max_price = price_stats[price_stats['genre'] == 'Fiction']['max_price'].values[0]
fiction_min_price = price_stats[price_stats['genre'] == 'Fiction']['min_price'].values[0]
non_fiction_max_price = price_stats[price_stats['genre'] == 'Non Fiction']['max_price'].values[0]
non_fiction_min_price = price_stats[price_stats['genre'] == 'Non Fiction']['min_price'].values[0]

print(f"Maximum price for Fiction: ${fiction_max_price}")
print(f"Minimum price for Fiction: ${fiction_min_price}")
print(f"Maximum price for Non Fiction: ${non_fiction_max_price}")
print(f"Minimum price for Non Fiction: ${non_fiction_min_price}")

books_count = df.groupby('author').agg(book_count=('name', 'count')).reset_index()

print(f"\nDimension of books_count table: {books_count.shape}")

most_books_author = books_count.loc[books_count['book_count'].idxmax()]
print(f"\nAuthor with the most books: {most_books_author['author']}")
print(f"Number of books by this author: {most_books_author['book_count']}")

average_ratings = df.groupby('author').agg(avg_rating=('user_rating', 'mean')).reset_index()

lowest_avg_rating_author = average_ratings.loc[average_ratings['avg_rating'].idxmin()]
print(f"\nAuthor with the lowest average rating: {lowest_avg_rating_author['author']}")
print(f"Average rating of this author: {lowest_avg_rating_author['avg_rating']}")

merged_df = pd.concat([books_count.set_index('author'), average_ratings.set_index('author')], axis=1)

sorted_merged_df = merged_df.sort_values(by=['book_count', 'avg_rating'], ascending=[True, True]).reset_index()

first_author_sorted = sorted_merged_df.iloc[0]
print(f"\nFirst author on the sorted list: {first_author_sorted['author']}")
print(f"Number of books: {first_author_sorted['book_count']}, Average rating: {first_author_sorted['avg_rating']}")
```
