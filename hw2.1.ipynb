```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'regions': ['Autonomous Republic of Crimea', 'Vinnytsia region', 'Volyn region', 'Dnipropetrovsk region', 'Donetsk region', 'Fatigue region', 'Transcarpathian regio', 'Zaporizhia region', 'Ivano-Frankivsk region', 'Kyiv region', 'Kirovohrad region', 'Luhansk region', 'Lviv region', 'Mykolayiv region', 'Odessa region', 'Poltava region', 'Rivne region', 'Sumy region', 'Ternopil region', 'Kharkiv region', 'Kherson region', 'Khmelnytsky region', 'Cherkasy regio', 'Chernivtsi region', 'Chernihiv region', 'city of Kyiv'],
    '2019': [None, 1.2, 1.48, 1.11, 1.11, 1.23, 1.52, 1.09, 1.28, 2.15, 1.12, 1.12, 1.26, 1.12, 1.33, 1.05, 1.53, 0.97, 1.12, 1.02, 1.27, 1.26, 1.03, 1.3, 1.03, 1.48]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10,6))
plt.bar(df['regions'], df['2019'], color='skyblue')
plt.title('Birth rate by regions in 2019', fontsize=16)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Birth rate (per 1,000)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```
