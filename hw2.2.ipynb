```python
import pandas as pd
path = '2017_jun_final - 2017_jun_final.csv'

try:
    df = pd.read_csv(path)
    print("File successfully loaded!")

    df.replace(" ", pd.NA, inplace=True)

    columns_to_keep = ["Programming language"]

    df_cleaned = df.dropna(axis=1, how='any')

    if "Programming language" not in df_cleaned.columns:
        df_cleaned = pd.concat([df_cleaned, df["Programming language"]], axis=1)

    null_proportion_cleaned = df_cleaned.isnull().mean() * 100
    print("\nProportion of missing values in the cleaned DataFrame:")
    print(null_proportion_cleaned)

    print("\nCleaned DataFrame:")
    print(df_cleaned.head())

except FileNotFoundError as e:
    print(f"Error: {e}")
```

```python
import pandas as pd
table = pd.read_csv('2017_jun_final - 2017_jun_final.csv')

numeric_columns = table.select_dtypes(include=['number'])

grouped_data = table.groupby('Role')[numeric_columns.columns].mean()
print("Grouped data by 'Role'(mean of numeric columns):")

display(grouped_data)
```

```python
import pandas as pd
table = pd.read_csv('2017_jun_final - 2017_jun_final.csv')

grouped_data = table.groupby('Role').agg({
        'Salary/month': ['min', 'max']
    })

grouped_data.columns = ['Min Salary', 'Max Salary']
print("Grouped data by 'Role' with min and max salary:")
    
display(grouped_data)
```
